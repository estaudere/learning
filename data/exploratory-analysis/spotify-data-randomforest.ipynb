```python
spotify_data = pd.read_csv('../input/spotify-dataset-19212020-160k-tracks/data.csv')
spotify_data.describe()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
```

```python
features = ['danceability', 'duration_ms', 'energy', 'instrumentalness', 'liveness', 'tempo']
X = spotify_data[features]
y = spotify_data.popularity

train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=1)
popularity_model = RandomForestRegressor(random_state=1)
```

```python
popularity_model.fit(train_X, train_y)
```

```python
pop_predictions = popularity_model.predict(val_X)
pop_predictions
```

```python
pop_mae = mean_absolute_error(pop_predictions, val_y)
print('Validation MAE for predicting Spotify popularity: {:,.0f}'.format(pop_mae))
```
